```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
krajDf = pd.read_excel("data/kraj_okres_link_absence.xlsx")
krajDf
```

```python
%%time
for index, krajRow in krajDf.iterrows():
    kraj = krajRow['Kraj']
    row_link = krajRow['Link']
    
    data = []
    url = f"{row_link}&search="
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")

    table = soup.find('table', attrs={'class':'list'})
    rows = table.find_all('tr')
    for row in rows[1:]:
        try:
            title_html = row.find('a')
            
            title = title_html.text.strip()
            doctor_url = ""
            if title_html.has_attr('href'):
                doctor_url = title_html['href']
            
            try:
                specialization = row.find_all('span')[0].text.replace("(", " ").replace(")", "")
                doctor_id = row.find_all('span')[1].text.replace("(", " ").replace(")", "")
            except:
                specialization = ""

            div = row.find('div').text.strip()

            data_row = {
                "title" : title,
                "doctor_url" : doctor_url,
                "specialization" : specialization,
                "kraj" : kraj,
                "nepritomnost" : div,
                "doctor_id" : doctor_id

            }
            data.append(data_row)
        except:
        # 
            pass
    df = pd.DataFrame(data)
    df.to_csv(f"data/kraj_absence_data/{kraj}_absence_doctor_list.csv",index=False)

```

```python
pd.read_csv("data/kraj_absence_data/Žilinský samosprávny kraj_absence_doctor_list.csv").to_csv("žilina_absence.csv", index=False)
```
